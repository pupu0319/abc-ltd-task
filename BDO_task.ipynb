## Import Employee Data and Payroll Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sheet1 = 'Emp List'
abc_1 = pd.read_excel('ABC Ltd.xlsx', sheet_name=sheet1)
abc_1
```

```python
new_header = abc_1.iloc[0]
abc_1 = abc_1[1:]
abc_1.columns = new_header
abc_1.reset_index(drop=True, inplace=True)
abc_1
```

```python
abc_1 = abc_1.drop(columns = 'Classification')
abc_1
```

```python
abc_1.rename(columns={'First Aid Certificate?': 'First Aid Certificate'}, inplace=True)
abc_1
```

```python
abc_1 = abc_1.fillna(value={"First Aid Certificate": "No"})
abc_1
```

```python
sheet3 = 'Payroll Data'
abc_3 = pd.read_excel('ABC Ltd.xlsx', sheet_name=sheet3)
abc_3
```

```python
new_header3 = abc_3.iloc[0]
abc_3 = abc_3[1:]
abc_3.columns = new_header3
abc_3.reset_index(drop=True, inplace=True)
abc_3
```

```python
new_abc_3 = abc_3.drop(abc_3.index[-1])
new_abc_3
```

```python
merged_file = abc_1.merge(new_abc_3, how='inner', on='Name', validate='many_to_one')
merged_file
```

```python
merged_file_cleaned = merged_file.dropna(axis=1)
merged_file_cleaned
```

```python
merged_file_cleaned = merged_file_cleaned.drop(columns = 'ID_y')
merged_file_cleaned.rename(columns={'ID_x': 'Employee ID'}, inplace=True)
merged_file_cleaned.rename(columns={'Name': 'Employee Name'}, inplace=True)
merged_file_cleaned.rename(columns={'Period ending': 'Date'}, inplace=True)
merged_file_cleaned
```

## Import Timesheet Data

```python
sheet2 = 'Timesheet'
abc_2 = pd.read_excel('ABC Ltd.xlsx', sheet_name=sheet2)
abc_2
```

```python
new_header = abc_2.iloc[0]
abc_2 = abc_2[1:]
abc_2.columns = new_header
abc_2.reset_index(drop=True, inplace=True)
abc_2
```

Merging timesheet with previous cleaned merged file

```python
merged = merged_file_cleaned.merge(abc_2, how='inner', on='Employee Name', validate='many_to_one')
merged
```
